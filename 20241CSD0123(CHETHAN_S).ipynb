## Deep Dive Case 1: Search vs Gen AI

### Scenario: Research 2025 AI Trends

#### Traditional Search Engine
* **Result:** List of 10+ blue links (Webpages).
* **User Action:** Click -> Read -> Go back -> Click -> Read -> Summarize manually.
* **Pain Point:** High cognitive load.

#### Generative AI (ChatGPT / Gemini)
* **Result:** A synthesized summary report.
* **User Action:** Read and verify.
* **Benefit:** Drastic reduction in information processing time.

## Deep Dive Case 2: Predict vs Create

### Scenario: Marketing Campaign Launch

#### Traditional AI (Machine Learning)
* **Task:** "Who will buy this product?"
* **Action:** Predicts churn rates, click-through probabilities, and customer segments.
* **Output:** Numbers, Scores, Lists.

#### Generative AI
* **Task:** "Make an ad for this product."
* **Action:** Writes email subject lines, generates ad images, writes blog posts.
* **Output:** Content (Text, Image).

```python
from sklearn.tree import DecisionTreeClassifier

# Data: [Age, Salary in $1k]
X = [[22, 30], [25, 45], [47, 110], [52, 90], [30, 70]]
y = [0, 0, 1, 1, 0]  # Output: 0 = No Buy, 1 = Buy

clf = DecisionTreeClassifier()
clf.fit(X, y)

prediction = clf.predict([[45, 100]])
print("--- Traditional ML Prediction ---")
print(f"Prediction for user [Age 45, Salary $100k]: {'Will Buy' if prediction[0] == 1 else 'Will Not Buy'}")
```

```python
import os
from google import genai
from google.colab import userdata

api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key)

print("=== Case 1: Generative Search & Synthesis ===")
res1 = client.models.generate_content(
    model='gemini-2.5-flash',
    contents="Summarize 3 key trends in AI for 2026 in short bullet points."
)
print(res1.text)

print("\n=== Case 2: Content Creation / Marketing Ad ===")
res2 = client.models.generate_content(
    model='gemini-2.5-flash',
    contents="Write a catchy 2-line marketing ad for Eco-Friendly Sneakers."
)
print(res2.text)
```
